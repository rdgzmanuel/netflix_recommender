```python
import pandas as pd
import re
```

```python
df = pd.read_csv("netflix1.csv")
df["Punctuation"] = 0
df
```

```python
run = True
while run:
    last_seen_title = input("Enter the title of the last film you've seen: ")
    try:
        last_seen = df.loc[df["title"] == last_seen_title]
        product = last_seen.iloc[0, 1]
        director = last_seen.iloc[0, 3]
        country = last_seen.iloc[0, 4]
        date_added = last_seen.iloc[0, 5]
        listed_in = last_seen.iloc[0, 9]
        run = False
    except:
        print("Film not found!!!")
matches = [0 for i in range(len(df))]

for index, row in df.iterrows():
    if re.findall(country, row["country"]) != []:
        matches[index] += 2
    if re.findall(director, row["director"]) != []:
        matches[index] += 10
    if re.findall(date_added, row["date_added"]) != []:
        matches[index] += 3
    if re.findall(listed_in, row["listed_in"]) != []:
        matches[index] += 8

df["Punctuation"] += matches
df_sorted = df.sort_values('Punctuation', ascending = False)

df_sorted.head()

recommended_df = df_sorted.iloc[1:10]    # Seleccionamos las 10 con mejor puntuaciÃ³n

print("You could try the following movies: ")
print(recommended_df.title)
```
